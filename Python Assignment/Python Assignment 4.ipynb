```python
# Q1. class to calucate the area
```

```python
class geometric:
    
    def __init__(self, a, b, c, s):
        self.a = a
        self.b = b
        self.c = c
        self.s = s

    def area(self):
        return (self.s*(self.s-self.a)*(self.s-self.b)*(self.s-self.c)) ** 0.5
```

```python
geo = geometric(10, 2, 3, 4)
geo.area()
```

```python
# Q2. function to return word longer than n
```

```python
def filter_long_words(arr, n=0):
    return [word for word in arr if len(word) > n]
filter_long_words('This is a sentences'.split(), 3)
```

```python
# Q3: return length of word
```

```python
def word_len(arr):
    return list(map(lambda x: len(x), arr))
word_len('ab,cde,erty'.split(','))
```

```python
# Q4: vowel/consonent
```

```python
def vowels(word):
    if word in 'a e i o u'.split():
        return True
    else:
        return False

vowels('s')
```
