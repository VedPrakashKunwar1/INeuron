```python
def divide(a, b):
    try:
        return(a/b)
    except ZeroDivisionError:
        return 'Zero divison is not possible'
print(divide(10, 5))
print(divide(10, 0))
```

```python
def sentence(subjects, verbs, objects):
    return [(s+" "+v+" "+o) for s in subjects for v in verbs for o in objects]
sentence(["Americans", "Indians"], ["Play", "watch"], ["Baseball","cricket"])
```
