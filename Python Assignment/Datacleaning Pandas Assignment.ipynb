```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
# Q1: Fill FlightNumber missing value and change data type to interger
```

```python
df['FlightNumber'] = df.loc[0, 'FlightNumber'] + 10*df.index
df['FlightNumber'] = df['FlightNumber'].astype('int')
df
```

```python
# Q2: split From_To to From and To column in temporary dataframe
```

```python
df1 = df
df1[['From', 'To']] = df1['From_To'].str.split('_', expand=True)
df1
```

```python
# Q3: captilise the from and to column
```

```python
df1.From = df1.From.str.capitalize()
df1.To = df1.To.str.capitalize()
df1
```

```python
# Q4: attach From and To column from temporary data frame 
```

```python
df
```

```python
df = pd.concat([df, df1], axis=1, join='inner' ).iloc[:, 7:]
df
```
